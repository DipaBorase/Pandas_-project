```python
import pandas as pd
```

```python
customers=pd.read_csv("/content/customers.csv",
                      header= None,
                      names=['custno',
                             'firstname',
                             'lastname',
                             'gender',
                             'age',
                             'profession',
                             'contactNo',
                             'emailId',
                             'city',
                             'state',
                             'isActive',
                             'createdDate',
                             'UpdatedDate'])
customers
```

```python
products=pd.read_csv("/content/products.csv",
                     header=None,
                     names=['productno',
                            'productName',
                            'Description',
                            'isActive',
                            'createdDate',
                            'UpdatedDate'])
products
```

```python
transactions=pd.read_csv("/content/transactions.csv",header=None,names=['txnno','txndate','custno','amount','productno','spendby'])
transactions
```

Top Five Records


```python
print(" Customers first five records")
customers.head()
```

```python
print(" Transactions first five records")
transactions.head()
```

```python
print("Products first five records")
products.head()
```

Customers purchased more than 3 products

```python
df=transactions.groupby('custno').count()
df1=df.sort_values(by='amount',ascending=False)
top_3=df1[df1>3].dropna()
pd.merge(customers,top_3,on='custno')
```

top 5 most demanded products

```python
df=transactions.groupby('custno').count()
df1=df.sort_values(by='amount',ascending=False).head()
pd.merge(products,df1,on='productno')
```

```python
transactions.sort_values(by='amount',ascending=False).head()
```

custno, gender, age, profession, contactno, productno, productName,
txndate, spendby, amount from CSV_s1 for custno = 923

```python
A=customers[customers['custno']==923]
df=pd.merge(A,transactions,on='custno')
df1=pd.merge(df,products,on='productno')
df1[['custno', 'gender', 'age', 'profession', 'productno', 'contactNo','productName', 'txndate', 'spendby', 'amount']]
```
